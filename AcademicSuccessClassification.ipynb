```python
from sklearn.model_selection import train_test_split

X = academic_data_train.iloc[:,:-1]
y = academic_data_train.iloc[:,-1]

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X, y)
```

```python
predictions = rfc.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, classification_report, confusion_matrix
print("ACCURACY OF THE MODEL:", accuracy_score(y_valid, predictions))
print("Classification Report:", classification_report(y_valid, predictions))
print("Confusion Matrix:", confusion_matrix(y_valid, predictions))
```

```python
test_data = pd.read_csv('/kaggle/input/playground-series-s4e6/test.csv')
test_predictions = rfc.predict(test_data)
test_data_prediction = pd.DataFrame({'Id':test_data.id, 'Target':test_predictions})
test_data_prediction.to_csv('submission.csv', index=False)
```
